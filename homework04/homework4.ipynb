## 4.3

Create the CNN

```python
from keras import layers
from keras import models
from keras.datasets import fashion_mnist
from keras.utils import to_categorical
from keras.regularizers import l2

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(28, 28, 1), activity_regularizer=l2(0.0001)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), padding='same', activation='relu', input_shape=(28, 28, 1), activity_regularizer=l2(0.0001)))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu', activity_regularizer=l2(0.0001)))
model.add(layers.Dense(10, activation='softmax'))
model.summary()
```

Train and test the model

```python
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

train_images = train_images.reshape((60000, 28, 28, 1))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 28, 28, 1))
test_images = test_images.astype('float32') / 255
#
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)

model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(train_images, train_labels, epochs=5, batch_size=64, validation_split=0.1)

test_loss, test_acc = model.evaluate(test_images, test_labels)
print(test_acc)
```

This model achieved 91.81% accuracy on the testing set. It achieved about 94.5% on the training set indicating
some overfitting. I added L2 regularization to the model in an attempt to combat the overfitting, but it still seems to occur.
